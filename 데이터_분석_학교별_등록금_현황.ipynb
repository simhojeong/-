# 원하는 대학교와 학과의 등록금을 비교

```python
# prompt: 대학교와 등록금을 넣으면 상위 대학교의 등록금의 차이를 비교하는 코드를 작성해줘.

import pandas as pd

# Sample data (replace with actual data)
data = {
    'University': ['SNU', 'Yonsei', 'Korea University', 'KAIST', 'POSTECH', 'University of Ulsan', 'GIST', 'Sungkyunkwan University', 'Hanyang University', 'Kyung Hee University'],
    'Tuition_Fee_KRW': [10000000, 9500000, 9000000, 8500000, 8000000, 7500000, 7000000, 6500000, 6000000, 5500000]
}

df = pd.DataFrame(data)

def compare_tuition(target_university, target_tuition):
    """
    Compares the tuition fee of a target university with other universities.

    Args:
        target_university: The name of the target university.
        target_tuition: The tuition fee of the target university.

    Returns:
        A DataFrame sorted by tuition fee difference, showing the difference between the target university's tuition and other universities' tuition fees.
        Returns an error message if the target university is not found.
    """

    if target_university not in df['University'].values:
        return f"Error: University '{target_university}' not found in the dataset."

    # Calculate the difference in tuition fees
    df['Tuition_Difference'] = df['Tuition_Fee_KRW'] - target_tuition

    # Sort the DataFrame by tuition difference
    df_sorted = df.sort_values('Tuition_Difference', ascending=False)

    return df_sorted

# Example usage
target_university = 'Yonsei'
target_tuition = df[df['University'] == target_university]['Tuition_Fee_KRW'].iloc[0]

comparison_result = compare_tuition(target_university, target_tuition)
comparison_result
```
